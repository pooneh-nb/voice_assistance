```python
import utilities
from subprocess import call
from subprocess import check_call
import os, sys
import tldextract
import pandas as pd
```

## load traffic

```python
base_out = "/home/c6/Desktop/single_pcaps"
pcap = "/home/c6/Desktop/single_pcaps/pandora.pcap"
```

## cnvert pcap to json

```python
jsonFile = pcap + "-ENC-out.json"
```

```python
cmd = ["tshark",
           "-o", "tcp.analyze_sequence_numbers:TRUE",
           "-o", "tcp.desegment_tcp_streams:TRUE",
           "-o", "http.desegment_body:TRUE",
           "-r", pcap, "-T", "json"]

with open(jsonFile, "wb") as jf:
    check_call(cmd, stdout=jf)
```

## DNS Resolutions

```python
dns_resolutions = {}
pcap_name = pcap.split('/')[-1].split('.')[0]


network_json_dir = os.path.join(base_out, pcap_name+".pcap-ENC-out.json")
network_json = utilities.read_json(network_json_dir)

for i in range(len(network_json)):
    if network_json[i]["_source"]["layers"]["frame"]["frame.protocols"] == 'eth:ethertype:ip:udp:dns':
        answers_key = []
        if "Answers" in network_json[i]["_source"]["layers"]["dns"].keys():
            answers_key  = network_json[i]["_source"]["layers"]["dns"]["Answers"].keys()
            for key in answers_key:
                if "CNAME" in key:
                    #print("CNAME")
                    continue
                try:    
                    dns_a = network_json[i]["_source"]["layers"]["dns"]["Answers"][key]["dns.a"]
                    dns_name = network_json[i]["_source"]["layers"]["dns"]["Answers"][key]["dns.resp.name"]
                    dns_resolutions[dns_a] = dns_name
                except:
                    pass
utilities.write_json(os.path.join(base_out, "dns_resolutions.json"), dns_resolutions)
```

## Un-unique Domains

```python
ip_list = []
for i in range(len(network_json)):
        if 'ip' in network_json[i]["_source"]["layers"].keys():
            if not network_json[i]["_source"]["layers"]['ip']['ip.src'].startswith('10.42') and not network_json[i]["_source"]["layers"]['ip']['ip.src'].startswith('255.255.'):
                ip_list.append(network_json[i]["_source"]["layers"]['ip']['ip.src'])
len(ip_list)
print(len(set(ip_list)))
```

### create a table of category | skill_name | FQDN | eSLD | subdomains |organization | ATSes | party

```python
dns_resolutions['54.239.18.66'] = "dss-na.amazon.com"
```

```python
dns_resolutions
```

```python
df = {'fqdn': [], 'domain_name': [], 'SLD': [], 'subdomain' : [] ,'organization': [], 'ATSes': [], 'Party': []}

jsons = os.path.join(base_out, pcap_name+".pcap-ENC-out.json")
ip_list = []
skill_network_json = utilities.read_json(jsons)

for i in range(len(skill_network_json)):
    if 'ip' in skill_network_json[i]["_source"]["layers"].keys():
        ip_src =  skill_network_json[i]["_source"]["layers"]['ip']['ip.src']
        ip_dst =   skill_network_json[i]["_source"]["layers"]['ip']['ip.dst']
        if not ip_src.startswith("10.42.") and not ip_src.startswith("255.255.") and not ip_src.startswith("8.8.") and not ip_src.startswith("0.0.") and not ip_src.startswith("224.0") and not ip_src.startswith("239.255"):
            if ip_src not in ip_list:
                #ip_list.(ip_src)
                if ip_src in dns_resolutions:
                    df['fqdn'].append(dns_resolutions[ip_src])
                    ext = tldextract.extract(dns_resolutions[ip_src])
                    df['domain_name'].append(ext.domain+"."+ext.suffix)
                    df['subdomain'].append(ext.subdomain)
                    df['SLD'].append(ext.domain)

                else:
                    df['fqdn'].append(ip_src)
                    df['domain_name'].append("")
                    df['subdomain'].append("")
                    df['SLD'].append("")
                df['organization'].append("")
                df['ATSes'].append("")
                df['Party'].append("")

        if not ip_dst.startswith("10.42.") and not ip_dst.startswith("255.255.") and not ip_dst.startswith("0.0.") and not ip_dst.startswith("8.8.") and not ip_dst.startswith("224.0") and not ip_dst.startswith("239.255"):
            if ip_dst not in ip_list:
                #ip_list.add(ip_dst)
                if ip_dst in dns_resolutions:
                    df['fqdn'].append(dns_resolutions[ip_dst])
                    ext = tldextract.extract(dns_resolutions[ip_dst])
                    df['domain_name'].append(ext.domain+"."+ext.suffix)
                    df['subdomain'].append(ext.subdomain)
                    df['SLD'].append(ext.domain)
                else:
                    df['fqdn'].append(ip_dst)
                    df['domain_name'].append("")
                    df['subdomain'].append("")
                    df['SLD'].append("")
                df['organization'].append("")
                df['ATSes'].append("")
                df['Party'].append("")
```

```python
df = pd.DataFrame.from_dict(df)
df.to_csv(os.path.join(base_out,"report.csv"), sep=',')
df
```

```python
unique_df = {}
for index, row in df.iterrows():
    if row['fqdn'] not in unique_df:
        unique_df[row['fqdn']] = {'domain_name': row['domain_name'], 'SLD': row['SLD'], 'subdomain' : row['subdomain'] ,'organization': "", 'ATSes': "", 'Party': ""}
```

## Address Organization

```python
ddg_entities = utilities.get_files_in_a_directory("/home/c6/Desktop/single_pcaps/tracker-radar/entities")

for fqdn, details in unique_df.items():
    
    for entity in ddg_entities:
        entity_content = utilities.read_json(entity)
        if details['domain_name'] in entity_content["properties"]:
            details['organization'] = entity_content["name"]
```

```python
non_org_in_ddg = {}
for fqdn, details in unique_df.items(): 
    if details['organization'] == '':
        if fqdn not in non_org_in_ddg:
            non_org_in_ddg[fqdn] = details['domain_name']
```
